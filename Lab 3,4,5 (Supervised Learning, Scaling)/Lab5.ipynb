```python
#load adult.csv into df_adult
df_adult = pd.read_csv('/content/drive/MyDrive/ML_Files/Lab5/adult.csv')

#load mushrooms.csv into df_mushroom
df_mushroom = pd.read_csv('/content/drive/MyDrive/ML_Files/Lab5/mushrooms.csv')
```

```python
#Scaling for adult.csv

# Standard Scaling (Z-score normalization)
scaler_standard = StandardScaler()
df_adult['standardized'] = scaler_standard.fit_transform(df_adult[['fnlwgt']])

# Min-Max Scaling (range 0–1)
scaler_minmax = MinMaxScaler()
df_adult['minmax'] = scaler_minmax.fit_transform(df_adult[['fnlwgt']])

# Robust Scaling (using median and IQR)
scaler_robust = RobustScaler()
df_adult['robust'] = scaler_robust.fit_transform(df_adult[['fnlwgt']])
```

```python
# Variance Threshold Filter Method for adult.csv

vt = VarianceThreshold(threshold=0.15)
vt.fit(df_adult._get_numeric_data()) # Apply to df_adult and only include numeric columns

result = pd.DataFrame({
    'variance': vt.variances_,
    'selected': vt.get_support()
}, index=df_adult._get_numeric_data().columns) # Update index to reflect numeric columns

print("Variance Threshold Results:\n", result)
```
