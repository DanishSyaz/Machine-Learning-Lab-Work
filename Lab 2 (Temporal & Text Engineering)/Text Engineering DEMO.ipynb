```python
#FEATURE ENGINEERING DEMO (Text data)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, PolynomialFeatures, PowerTransformer
from sklearn.metrics import mean_squared_error, accuracy_score

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import re, nltk

```

```python
# Download NLTK resources for tokenization, stopwords, and lemmatization
nltk.download("punkt")        # For tokenization
nltk.download("punkt_tab")    # Extra fix for some environments (LookupError)
nltk.download("stopwords")    # Stopwords list
nltk.download("wordnet")      # WordNet for lemmatization

# Load stopwords and lemmatizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words("english"))  # English stopword set
lemmatizer = WordNetLemmatizer()              # Lemmatizer object

# TEXT CLEANING FUNCTION
def clean_text(s):
    # Convert text to lowercase and remove non-alphabetic characters
    s = re.sub(r"[^a-zA-Z]", " ", s.lower())

    # Tokenize the string into words
    tokens = nltk.word_tokenize(s)

    # Remove stopwords, short words, and lemmatize remaining tokens
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop_words and len(t) > 2]

    # Join back into a single cleaned string
    return " ".join(tokens)
```

```python
from pathlib import Path

# Put folder path in google drive that contains all .txt file
folder_path = Path("/content/drive/MyDrive/ML_Files/Text_Engineering/")

# store all file contents
data = []

#loop through the folder to append all .txt data
for file_path in folder_path.glob("*.txt"):
    if file_path.is_file():
        with open(file_path, encoding = "ISO-8859-1") as f:
            content = f.read()
            data.append(content)
```

```python

# Clean first 200 documents for faster demo
texts = [clean_text(t) for t in data[:200]]

# BAG OF WORDS (BoW)

# Convert text into numeric features using word counts
cv = CountVectorizer(max_features=1000)  # Keep top 1000 frequent words
X_bow = cv.fit_transform(texts)
print("BoW shape:", X_bow.shape)

# TF-IDF (Term Frequency - Inverse Document Frequency)

# Converts text into weighted word features
tfidf = TfidfVectorizer(max_features=1000, ngram_range=(1, 2))  # Use unigrams + bigrams
X_tfidf = tfidf.fit_transform(texts)
print("TF-IDF shape:", X_tfidf.shape)

# TOPIC MODELING WITH LDA (Latent Dirichlet Allocation)

 # Extract 5 topics
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(X_bow)

print("LDA Topics:")
terms = cv.get_feature_names_out()

# Print top 10 words for each topic
for idx, comp in enumerate(lda.components_):
    top_terms = [terms[i] for i in comp.argsort()[-10:]]
    print("Topic", idx, ":", " ".join(top_terms))
```
